```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
Book_reader = Bookclean.pivot(index='User',columns='Book',values='Rating').reset_index(drop=True)
Book_reader
```

```python
Book_reader.fillna(0,inplace=True)
Book_reader
```

```python
Book_sim = 1-pairwise_distances(Book_reader.values,metric='cosine')
Book_sim
```

```python
Book_sim_df=pd.DataFrame(Book_sim)
```

```python
Book_sim_df.index=Bookclean.User.unique()
Book_sim_df.colums=Bookclean.User.unique()
```
