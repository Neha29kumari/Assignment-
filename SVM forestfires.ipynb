# Creat the function categorycal to numerical data


```python
from sklearn.preprocessing import LabelEncoder
```

```python
def object_to_int(Data):
    if Data.dtype == "object":
        Data = LabelEncoder().fit_transform(Data)
    return Data
```

```python
Data = Data.apply(lambda x: object_to_int(x))
Data.head()
```

```python
Data.tail()
```

```python
Data.describe(include='all')
```

```python
x=Data.iloc[:,:-1]
y=Data.iloc[:,-1]
```

```python
x
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=40)
```

```python
print(x_test.shape,x_train.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

# Applying support Vector Machine for classification

```python
from sklearn.svm import SVC
classifire = SVC(kernel='linear',random_state=0)
classifire.fit(x_train,y_train)
```

```python
#predicting the test set rejults
y_pred = classifire.predict(x_test)
```

```python
y_pred
```

```python
from sklearn import metrics
print('Accuracy score with linear kernal')

print(metrics.accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'rbf')
classifier.fit(x_train, y_train)

y_pred=classifier.predict(x_test)

print('Accuracy score  with default rbf kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
classifier = SVC(kernel = 'rbf', gamma=12, C=2,random_state=0)
classifier.fit(x_train, y_train)

y_pred=classifier.predict(x_test)

print('Accuracy score on test data with default rbf linear kernal')
print(metrics.accuracy_score(y_test,y_pred))
```

```python
svc=SVC(kernel='poly',degree=4)
svc.fit(x_train,y_train)

y_pred=svc.predict(x_test)
print('Accuracy score with poly kernal')
print(metrics.accuracy_score(y_test,y_pred))
```
